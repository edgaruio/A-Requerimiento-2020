### Consulta masiva

```python
n = 23
datos_final = pd.DataFrame() # Contenedor vacio
for i in range(1,n+1):
    url_i = "https://colsubsidio.vtexcommercestable.com.br/api/oms/pvt/orders?page="+str(i)+"&per_page=100"
    response_i = requests.request("GET", url_i, headers=headers, params=querystring)
    datos_i = response_i.json()
    dict_list_i = next(iter(datos_i.values()))
    datos_final = datos_final.append(pd.DataFrame(dict_list_i))
```
